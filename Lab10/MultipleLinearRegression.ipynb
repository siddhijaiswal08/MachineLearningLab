```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X=np.array([
    [1,1],
    [2,1],
    [3,2],
    [4,3],
    [5,3],
])
y=np.array([20,40,50,65,80])
```

```python
model=LinearRegression()
model.fit(X,y)
```

```python
y_pred=model.predict(X)
```

```python
residuals=y-y_pred
```

```python
mae=mean_absolute_error(y,y_pred)
mse=mean_squared_error(y,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y,y_pred)
```

```python
print("Coefficient (b1=Hours, b2=Practice_Tests):",model.coef_)
```

```python
print("Intercept(b0):",model.intercept_)
```

```python
print("Mean Absolute Error:",mae)
print("Mean Squared Error:",mse)
print("Root Mean Squared Error:",rmse)
print("R2 score:",r2)
```

```python
plt.scatter(y_pred, residuals, color="blue",label="Residuals")
plt.axhline(y=0,color="red",linestyle="--")
plt.xlabel("residuals (Actual-Predicted)")
plt.ylabel("Residuals Analysis - Multiple Linear Regression (Student Scores)")
plt.legend()
plt.show()
```
